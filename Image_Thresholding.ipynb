```python
#Resultant image will only have black and white color
#works well on binary images
%matplotlib inline
import matplotlib.pyplot as plt
import skimage.data as data

img = data.camera()

#positive thresholding
#here we are manually setting the value of threshold
thresh =127
#kept middle value 

output1 = img > thresh
output2 = img <= thresh

output = [img,output1,output2]
titles=['Original Image','Thresholded','Inverted Thresolding']

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(output[i],cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()    

#thresholding divides image into two regions ====one black,other white
#fundamental form of image segmentation
```

```python
#Other forms of thresholding
#Bimodel histogram
#this algorithm calculates value of threshold from histogram
#otsu binarization

from skimage.filters import threshold_otsu
img = data.camera()
thresh = threshold_otsu(img)
#will automatically calculate value of threshold
#foreground clearly distinguishable from background 

#there are two peaks in histogram hence the name

binary = img > thresh

plt.subplots_adjust(wspace=0.5,hspace=0.5)

plt.subplot(1,3,1)
plt.title('Original Image')
plt.axis('off')
plt.imshow(img,cmap='gray')

plt.subplot(1,3,2)
plt.title('Bimodel Histogram')
plt.hist(img.ravel(),bins=256,range=[0,255])
plt.xlim([0,255])

plt.subplot(1,3,3)
plt.title('Thresholded Image')
plt.axis('off')
plt.imshow(binary,cmap='gray')

plt.show()

#otsu algorithm works when image is in bimodel format i.e background foreground can be easily distinguished
```

```python
#adaptive thresholding
#images where background and foreground cannot be easily distinguishable
#comaparing otsu with adaptive

from skimage.filters import threshold_local
img=data.page()
thresh = threshold_otsu(img)
otsu=img > thresh

adaptive = threshold_local(img,block_size =5,offset=40)

output = [img,otsu,adaptive]
titles=['Original Image','Otsu Thresholding','Adptive Thresolding']

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(output[i],cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
