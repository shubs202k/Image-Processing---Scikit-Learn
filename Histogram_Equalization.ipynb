```python
#We enhance the tones of an image using histogram equalization
#we can adjust the contrast of an image by HE
%matplotlib inline
import matplotlib.pyplot as plt 
from skimage import data
from skimage import exposure

img=data.moon()

img_eq = exposure.equalize_hist(img)
img_adapthist = exposure.equalize_adapthist(img,clip_limit=0.03)
output = [img,img_eq,img_adapthist]
titles = ['Original','Histogram Equalization','Adaptive Equalization']

for i in range(3):
    plt.subplot(3,1,i+1)
    plt.imshow(output[i],cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()    

#image enhanced after HE
```
