```python
#Pixels with same intensity levels are clustered together and given same color
#Simple linear Iterative Clustering
%matplotlib inline
from skimage import data,segmentation,color
from skimage.future import graph
from matplotlib import pyplot as plt

img = data.coffee()

labels = segmentation.slic(img,compactness=30,n_segments=1500)
#clarity reduced with n-segments 100
#image with n_segments will have more information
out = color.label2rgb(labels,img,kind='avg')

plt.imshow(out)
plt.axis('off')
plt.show()
```
