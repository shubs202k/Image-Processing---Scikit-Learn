```python
%matplotlib inline

import matplotlib.pyplot as plt
from skimage import data, color, img_as_float

grayscale_image = img_as_float(data.moon()[::2, ::2])
image = color.gray2rgb(grayscale_image)
#Grayscale value representated as RGB,2-D grayscale representation to 3D Grayscale
#Not possible to add a tint of any color to a grayscale image

plt.imshow(image)
plt.show()
```

```python
#Adding red tint to original image
red_multiplier = [1, 0, 0]
output1 = image * red_multiplier
plt.imshow(output1)
plt.show()
```

```python
#Adding green tint to original image
green_multiplier = [0, 1, 0]
output2 = image * green_multiplier
plt.imshow(output2)
plt.show()
```

```python
#Adding blue tint to original image
blue_multiplier = [0, 0, 1]
output3 = image * blue_multiplier
plt.imshow(output3)
plt.show()
```

```python
#Adding yellow tint
yellow_multiplier = [1, 1, 0]
output4 = image * yellow_multiplier
plt.imshow(output4)
plt.show()
```

```python
#Adding cyan tint
cyan_multiplier = [0, 1, 1]
output4 = image * cyan_multiplier
plt.imshow(output4)
plt.show()
```
